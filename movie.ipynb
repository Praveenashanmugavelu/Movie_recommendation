```python
%%writefile app.py
import pandas as pd
import numpy as np
import requests
import streamlit as st
import pickle
import os
import base64

def fetch_movie_image(movie_id):
    url = "https://api.themoviedb.org/3/movie/{}?api_key=07317ce37d243da893b77e9120d87a86".format(movie_id)
    data = requests.get(url)
    data = data.json()
    poster_path = data['poster_path']
    full_path = "https://image.tmdb.org/t/p/w500/" + poster_path
    return full_path

def recommend_movie(movie):
    ind = movies[movies['title'] == movie].index[0]
    dist = sorted(list(enumerate(sim[ind])), reverse=True, key=lambda x: x[1])
    recom_movie_name = []
    recom_movie_image = []
    for i in dist[1:6]:
        movie_id = movies.iloc[i[0]].movie_id
        recom_movie_image.append(fetch_movie_image(movie_id))
        recom_movie_name.append(movies.iloc[i[0]].title)

    return recom_movie_name,recom_movie_image

st.header('Movie Recommender System')
movies=pickle.load(open('movie_dict.pkl','rb'))
sim = pickle.load(open('sim.pkl','rb'))

movies_dict = movies['title'].values
chosed_movie = st.selectbox("Type/select a movie to recommend",movies_dict)

if st.button('Recommend'):
    recom_movie_name,recom_movie_image = recommend_movie(chosed_movie)
    i1, i2, i3, i4, i5 = st.columns(5)
    with i1:
        st.image(recom_movie_image[0])
        st.text(recom_movie_name[0])
    with i2:
        st.image(recom_movie_image[1])
        st.text(recom_movie_name[1])
    with i3:
        st.image(recom_movie_image[2])
        st.text(recom_movie_name[2])
    with i4:
        st.image(recom_movie_image[3])
        st.text(recom_movie_name[3])
    with i5:
        st.image(recom_movie_image[4])
        st.text(recom_movie_name[4])
```

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/drive/MyDrive/tmdb_5000_credits.csv')
movies.head(2)
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.info()
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies=movies.drop_duplicates()
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head(1)
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head(1)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.sample(5)
```

```python
def colapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(colapse)
movies['crew'] = movies['crew'].apply(colapse)
movies['genres'] = movies['genres'].apply(colapse)
movies['keywords'] = movies['keywords'].apply(colapse)
```

```python
movies.head(1)
```

```python
movies['tag'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_df = movies.drop(columns=['overview','genres','keywords','cast','crew'])
new_df['tag'] = new_df['tag'].apply(lambda x: " ".join(x))
```

```python
new_df.head(1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vec = cv.fit_transform(new_df['tag']).toarray()
vec
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
p=PorterStemmer()
def stem(text):
  a=[]
  for i in text.split():
    a.append(p.stem(i))

  return " ".join(a)
```

```python
new_df['tag']=new_df['tag'].apply(stem)
```

```python
new_df['tag'][0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim=cosine_similarity(vec)
sorted(list(enumerate(sim[0])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recomend(movie):
  ind=new_df[new_df['title']==movie].index[0]
  dist=sorted(list(enumerate(sim[ind])),reverse=True,key=lambda x:x[1])
  for i in dist[1:6]:
    print(new_df.iloc[i[0]].title)
```

```python
recomend('Avatar')
```
